Now let us construct a Liouvillian numerically. Consider the Lindblad equation for a qubit in a thermal bath $$ \frac{{\rm d}}{{\rm d}t} \hat{\rho} = \frac{1}{{\rm i}\hbar} [ \hat{H},\hat{\rho}] + \gamma (1+\bar{n})\mathcal{D}[\hat{\sigma}_-]\hat{\rho} + \gamma \bar{n} \mathcal{D}[\hat{\sigma}_+]\hat{\rho} = \mathcal{L}\hat{\rho},$$ where $\hat{H} = \hbar\omega_0 \hat{\sigma}_z/2$, $\bar{n} = ({\rm e}^{\beta\hbar\omega_0}-1)^{-1}$ and $\beta = 1/k_BT$ is the inverse temperature. Take the parameters to be $\omega_0=1$, $\gamma = 0.1\omega_0$ and $k_BT = 0.5\hbar\omega_0$.

(a) The Pauli operator basis is defined by $\hat{e}_j = \hat{\sigma}_j/\sqrt{2}$, where $\{\hat{\sigma}_0,\hat{\sigma}_1,\hat{\sigma}_2,\hat{\sigma}_3\}= \{\hat{\mathbb{1}},\hat{\sigma}_x,\hat{\sigma}_y,\hat{\sigma}_z\}$. Verify numerically that this forms an orthonormal basis with respect to the inner product $\langle \hat{e}_i,\hat{e}_j\rangle = {\rm tr}(\hat{e}_i^\dagger \hat{e}_j)$.

```python
import numpy as np

I = np.array([[1,0], [0,1]])
X = np.array([[0,1], [1,0]])
Y = np.array([[0,-1j], [1j,0]])
Z = np.array([[1,0], [0,-1]])

paulis = [I, X, Y, Z]
e = [1/np.sqrt(2) * i for i in paulis]

#inner product function
def ip(a, b): 
    return np.trace(a.T.conjugate() @ b)
inner_prods = np.zeros((len(e), len(e)), dtype=complex) #matrix to store all the inner products values
for i in range(len(e)):
    for j in range(len(e)):
        inner_prods[i,j] = ip(e[i],e[j])
#if the inner product matrix is the identity, the basis is orthonormal
print(inner_prods)
if np.allclose(inner_prods, np.eye(len(e), len(e))):
    print('It is an orthonormal basis:D')
```

(b) Construct and print the Liouvillian operator $\hat{\mathcal{L}}$ as a NumPy array in the Pauli basis.

```python
w0 = 1
g = 0.1 * w0
hbar_w0_b = 1/0.5 #hbar * w0 * beta

I = np.array([[1,0], [0,1]])
X = np.array([[0,1], [1,0]])
Y = np.array([[0,-1j], [1j,0]])
Z = np.array([[1,0], [0,-1]])

paulis = [I, X, Y, Z]
e = [1/np.sqrt(2) * i for i in paulis]

H_hbar = w0 / 2 * Z
nbar = (np.exp(hbar_w0_b)-1)**(-1)

zero = np.array([[1],[0]])
one = np.array([[0], [1]])

sigma_plus = zero @ one.T
sigma_minus = one @ zero.T

#commutator function
def comm(a, b):
    return a @ b - b @ a
#anticommutator function
def anticomm(a, b):
    return a @ b + b @ a
#dephasing superoperator
def dephasing(L, rho):
    Ldag = L.T.conjugate()
    return L @ rho @ Ldag - 1/2 * anticomm(Ldag @ L, rho)
#Liouvillian superoperator
def li_superop(H_hbar, g, nbar, L1, L2, rho):
    return -1j * comm(H_hbar, rho) + g * (1 + nbar) * dephasing(L1, rho) + g * nbar * dephasing(L2, rho)

#constructing liouvillian in pauli basis
size = len(e)
L = np.zeros((size, size), dtype=complex)
for i in range(size):
    for j in range(size):
        L[i,j] = np.trace(e[i] @ li_superop(H_hbar, g, nbar, sigma_minus, sigma_plus, e[j])) #no need to take hermitian conjugate of i since Pauli matrices are hermitian
print(L)
```

(c) Find the eigenvalues and right eigenvectors of $\hat{\mathcal{L}}$. Print the numerically obtained eigenvectors in their standard matrix form (i.e. expressed in the computational basis $|0\rangle$ and $|1\rangle$). Verify that the steady state is thermal as expected. Comment on the implications of these results for the time evolution of the qubit.

```python
eigvals, eigvects = np.linalg.eig(L)
    
#computational basis
zero = np.array([[1],[0]])
one = np.array([[0], [1]])

p1 = zero @ zero.T
p2 = zero @ one.T
p3 = one @ zero.T
p4 = one @ one.T

p = [p1, p2, p3, p4]

#matrix for basis change
S = np.zeros((4,4), dtype=complex)
for i in range(4):
    for j in range(4):
        S[i,j] = np.trace( p[i].T @ e[j])

eigvects_comp = S @ eigvects

for i in range(len(eigvals)):
    print('*eigenval: ',eigvals[i])
    print('\t pauli basis')
    print(eigvects[:,i])
    print('\t computational basis')
    print(eigvects_comp[:,i])
```

```python
#to verify the state is thermal, we have to compute the liouvillian on the density matrix (i.e d/dt(rho))
for i in range(len(eigvals)):
    v = eigvects_comp[:, i]
    rho_t = v.reshape(2,2)
    liouvillian = li_superop(H_hbar, g, nbar, sigma_minus, sigma_plus, rho_t)
    if np.allclose(liouvillian, np.zeros(liouvillian.shape)):
        print(v)
        print('This is a thermal state:\n', rho_t, '\n with Liouvillian eigenvalue of', eigvals[i])
```

```python
def vectoriser(A, datatype = complex):
    v = np.zeros(len(A)**2, dtype=datatype)
    for i in range(len(A)):
        for j in range(len(A)):
            v[(len(A))*i + j] = A[i,j]
    return v
zero = np.array([[1],[0]])
one = np.array([[0], [1]])
psi = 1/np.sqrt(2) * (zero + 1j * one)
rho0 = psi @ psi.T.conjugate()
r0 = vectoriser(rho0) #vectorised state

y = vectoriser(Y)

if np.allclose(y.T.conjugate() @ r0, +1):
    print('The expectation value is +1 as expected')
```

(b) Write functions that generate the matrix representation of the left- and right-multiplication superoperators, $\hat{\mathcal{M}}_L[\hat{A}]$ and $\hat{\mathcal{M}}_R[\hat{A}]$, for an arbitrary operator $\hat{A}$ (see question 3). Use these functions to construct the Hamiltonian superoperator $\mathcal{H} = [\hat{H},\bullet]$ for a qubit with $\hat{H} = \hat{\sigma}_z/2$, and verify that $\mathcal{H}$ is Hermitian. 

```python
def ML(A):
    return np.kron(A, np.eye(len(A)))
def MR(A):
    return np.kron(np.eye(len(A)), A.T)

H = Z/2

def H_superop(H):
    return ML(H)-MR(H)
if np.allclose(H_superop(H), H_superop(H).T.conjugate()):
    print('The Hamiltonian superoperator is Hermitian')
```

Now let us simulate the quantum-jump trajectories for the master equation of question 6 describing a driven-dissipative qubit. Recall that this method involves simulating the dynamics of an *unnormalised* pure state $$ |\tilde{\psi}(t)\rangle = {\rm e}^{-{\rm i}\hat{H}_{\rm eff}t/\hbar}|\psi(0)\rangle,$$ where  the non-Hermitian Hamiltonian is $$ \hat{H}_{\rm eff} = \hat{H} -\frac{{\rm i \hbar\gamma }}{2} \hat{\sigma}_+\hat{\sigma}_-.$$ The waiting time distribution describing photon emissions, $W(t) = -{\rm d}P_0/{\rm d} t$, can then be sampled using the expression for the survival probability, $P_0(t) = \langle \tilde{\psi}(t)|\tilde{\psi}(t)\rangle.$

(a) As a warm-up, set $\Delta = \Omega = 0$ and consider the initial state $\hat{\rho}(0) = |0\rangle\langle 0| = |\uparrow\rangle\langle \uparrow|$. Derive an analytical expression for the waiting-time distribution for photon emissions.



(b) Numerically sample the waiting-time distribution by evolving the initial state under the non-Hermitian Hamiltonian, discretising time in small steps $\Delta t$ such that
$$ |\tilde{\psi}(t+\Delta t)\rangle = {\rm e}^{-{\rm i}\hat{H}_{\rm eff}\Delta t /\hbar}|\tilde{\psi}(t)\rangle.$$
Plot a histogram of the numerically obtained $W(t)$ as a function of $\gamma t$ for 200 trajectories, and quantitatively compare it with your analytical result. 

```python
psi = np.array([[1], [0]])

O = 0
d = 0
g = 1
H_hbar = d/2 * Z + O/2 * X
H_eff_hbar = H_hbar - 1j * g / 2 * np.array([[1, 0], [0, 0]])

time_size = 500
ti = 0
tf = 7.5/g
t = np.linspace(ti, tf, time_size)
dt = (tf-ti)/len(t)
gt = [g*i for i in t]

def simulation(psi0, Heff, t):
    r = np.random.random()
    U = expm(-1j * Heff * dt)
    waiting_times = []
    state = psi0
    for i in range(len(t)):
        state = U @ state
        P0 = state.T.conjugate() @ state
        if np.isclose(P0, 1-r, 1e-02):
            waiting_times.append(t[i])
            r = np.random.random()
    return waiting_times
distr = []
for i in range(200):
    s = simulation(psi, H_eff_hbar, t)
    distr += s

plt.figure(figsize=(11,4), dpi=300)
plt.hist(distr, bins=10, density = True)
plt.plot(gt, [g * np.exp(-g*i) for i in t], label=r'$W(t) = \gamma e^{-\gamma t}$')
plt.ylabel('Normalised # of waiting times')
plt.xlabel(r'$\gamma t$')
plt.legend()
plt.show()
```

Consider the anharmonic oscillator with Hamiltonian given in the interaction picture by $$ \hat{H} = \hbar\Delta \hat{a}^\dagger\hat{a} + \frac{\hbar U}{2}\hat{a}^\dagger \hat{a}^\dagger \hat{a}\hat{a} + \frac{\hbar G}{2} \left(\hat{a}^2 + \left(\hat{a}^\dagger\right)^2\right).$$ This describes a cavity mode interacting with a non-linear medium, where $\Delta$ is the detuning of the drive from resonance, $U$ quantifies the effect of the non-linear medium and $G$ is the strength of a driving field (the rotating wave approximation has been invoked). The system undergoes dissipation described by the Lindblad equation $$\frac{\rm d}{{\rm d}t} \hat{\rho} = \frac{1}{{\rm i}\hbar} [\hat{H},\hat{\rho}] + \gamma\mathcal{D}[\hat{a}]\hat{\rho}.$$ Find the steady state numerically and make a contour plot of its Husimi Q-function for $U = \gamma/3$ and $G=\gamma$, and with three values of the detuning: $\Delta/\gamma =-2,0,+2$. Comment briefly on the nature of the steady state in each case.

```python
%%time
g = 1
U = g/3
G = g
d = [-2*g, 0*g, 2*g]

def fock(number, d):
    f = np.zeros((1, d), dtype=complex)
    f[0, number] = 1
    return f.T

def anihilation(d):
    a = np.zeros((d, d), dtype=complex)
    for i in range(d-1):
        a[i,i+1] = np.sqrt(i+1)
    return a
dimension = 20
a = anihilation(dimension)
adag = a.T.conjugate()

H_hbar = [i * adag @ a + U/2 * adag @ adag @ a @ a + G/2 * (a @ a + adag @ adag) for i in d] #list of hamiltonians/hbar for every detunning

#liouvillian basis
focks = [fock(i, dimension) for i in range(dimension)]
basis = []
for i in focks:
    for j in focks:
        basis.append(i @ j.T.conjugate())

Liouvillian = []
thermal_states = []
for k in range(len(H_hbar)):
    L = np.zeros((dimension**2, dimension**2), dtype=complex)
    for i in range(dimension**2):
        for j in range(dimension**2):
            L[i,j] = np.trace(basis[i].T.conjugate() @ li_superop(H_hbar[k], g, a, basis[j])) #no need to take hermitian conjugate of i since the basis is real
    Liouvillian.append(L)
    eigvals, eigvects = np.linalg.eig(L)
    '''for i in range(len(eigvals)):
        print('*eigenval: ',eigvals[i])
        print('\t eigenvect')
        print(eigvects[:,i])'''

    #to verify the state is thermal, we have to compute the liouvillian on the density matrix (i.e d/dt(rho))
    for i in range(len(eigvals)):
        v = eigvects[:, i]
        rho = v.reshape(dimension,dimension)
        liouv = li_superop(H_hbar[k], g, a, rho)
        if np.allclose(liouv, np.zeros(liouv.shape)):
            #print('This is a thermal state:\n', rho, '\n with Liouvillian eigenvalue of', eigvals[i], 'for the driving field with detunning', d[k])
            thermal_states.append(rho)

        
def displacement(alpha, d):
    a = anihilation(d)
    D = expm(alpha * a.T.conjugate() - alpha.conjugate() * a)
    return D

#a function to calculate the Q-function at a specific point of the phase space given a state rho
def Qdiscr(alpha, rho, d):
    D = displacement(alpha, d)
    alpha_state = D @ fock(0, d) #D(alpha)|0>=|alpha>
    return np.real(1/np.pi * alpha_state.T.conjugate() @ rho @ alpha_state)

#a function that creates a phase space, and plots the Q-function value at every point of it
def Q_plot(rho, d, label):
    space_size = 100
    x = np.linspace(-5, 5, space_size)
    y = np.linspace(-5, 5, space_size)
    xx, yy = np.meshgrid(x, y)

    z1 = np.zeros((space_size, space_size))
    for i in range(space_size):
        for k in range(space_size):
            z1[i,k] = Qdiscr(x[i]+1j*y[k], rho, d)

    fig = plt.figure(dpi=100)

    ax1= fig.add_axes([0, 0, 0.8, 1])
    cont1 = ax1.contourf(xx, yy, z1, cmap = 'magma')
    ax1.set_title(r'$\Delta=$'+str(label)+'$\gamma$')
    ax1.set_xlabel(r'Re{$\alpha$}')
    ax1.set_ylabel(r'Im{$\alpha$}')
    ax1.grid(alpha=.2)

    cbar_ax = fig.add_axes([0.85, 0.1, 0.02, 0.8])
    fig.colorbar(cont1, cax=cbar_ax)
    plt.show()
    return x, y, z1
    
for i in range(len(thermal_states)):
    x, y, Q = Q_plot(thermal_states[i], dimension, d[i])        
        
```

We can see that the different peaks do not interfere so it is a mixed state. In particular, we have a mix of 3, 2 and 1 slightly squeezed states for $\Delta=-2$, $0$ and $+2$ respectively.
